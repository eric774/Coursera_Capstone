### Now, let's get the top 150 venues that are within a radius of 10000 meters.

```python
radius = 10000
LIMIT = 150

venues = []

for lat, long, neighborhood in zip(df2['Latitude'], df2['Longitude'], df2['City']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['City', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df.head()
```

## Analyze Each Neighborhood

```python
# one hot encoding
j_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add City column back to dataframe
j_onehot['City'] = venues_df['City'] 

# move City column to the first column
fixed_columns = [j_onehot.columns[-1]] + list(j_onehot.columns[:-1])
j_onehot = j_onehot[fixed_columns]

print(j_onehot.shape)
j_onehot.head()
```

```python
# Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
j_grouped = j_onehot.groupby(["City"]).mean().reset_index()

print(j_grouped.shape)
j_grouped
```

```python
len(j_grouped[j_grouped["Hostel"] > 0])
```

```python
# Create a new DataFrame for Hostel data only
j_hostel = j_grouped[["City","Hostel"]]
j_hostel.head()
```

## Cluster Neighborhoods

```python
# set number of clusters
kclusters = 3

j_clustering = j_hostel.drop(["City"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(j_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each City.
j_merged = j_hostel.copy()

# add clustering labels
j_merged["Cluster Labels"] = kmeans.labels_
j_merged.head()
```

```python
# merge data to add latitude/longitude for each City
j_merged = j_merged.join(df2.set_index("City"), on="City")

print(j_merged.shape)
j_merged.head() # check the last columns!
```

```python
# sort the results by Cluster Labels
print(j_merged.shape)
j_merged.sort_values(["Cluster Labels"], inplace=True)
j_merged
```

### Visualizing Clusters

```python
# Finally, let's visualize the resulting clusters

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(j_merged['Latitude'], j_merged['Longitude'], j_merged['City'], j_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-3],
        fill=True,
        fill_color=rainbow[cluster-2],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
